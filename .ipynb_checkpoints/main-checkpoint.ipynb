```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords if you haven't already
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# ==========================================
# 1. LOAD THE DATASET
# ==========================================
# Assuming 'flipkart.csv' is in the same directory
df = pd.read_csv('flipkart.csv')

print("--- Dataset Summary ---")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())

# ==========================================
# 2. DATA CLEANING & PREPROCESSING
# ==========================================
def clean_review(text):
    if not isinstance(text, str):
        return ""
    # Convert to lowercase
    text = text.lower()
    # Remove HTML tags, special characters, and numbers
    text = re.sub(r'<.*?>|[^a-zA-Z\s]', '', text)
    # Remove stopwords
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)

# Drop rows where 'Review' or 'Rating' is missing
df = df.dropna(subset=['Review', 'Rating'])

# Apply text cleaning
df['Cleaned_Review'] = df['Review'].apply(clean_review)

# ==========================================
# 3. SENTIMENT ANALYSIS
# ==========================================
def get_sentiment(text):
    analysis = TextBlob(text)
    # polarity score ranges from -1 (Negative) to +1 (Positive)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis
df['Sentiment'] = df['Cleaned_Review'].apply(get_sentiment)

print("\n--- Sentiment Analysis Distribution ---")
print(df['Sentiment'].value_counts())

# ==========================================
# 4. VISUALIZATIONS
# ==========================================
plt.figure(figsize=(14, 5))

# Plot 1: Sentiment Distribution (Bar Chart)
plt.subplot(1, 2, 1)
# Adding hue='Sentiment' and legend=False satisfies the new Seaborn requirements
sns.countplot(data=df, x='Sentiment', hue='Sentiment', palette='Set2', order=['Positive', 'Neutral', 'Negative'], legend=False)
plt.title('Overall Customer Sentiment')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Reviews')

# Plot 2: Sentiment vs Rating (Stacked Bar / Cross-tabulation)
plt.subplot(1, 2, 2)
# Create a cross-tabulation of Ratings and Sentiments
sentiment_rating = pd.crosstab(df['Rating'], df['Sentiment'])
# Ensure order of columns for consistency
sentiment_rating = sentiment_rating.reindex(columns=['Positive', 'Neutral', 'Negative'], fill_value=0)

sentiment_rating.plot(kind='bar', stacked=True, color=['#2ca02c', '#9467bd', '#d62728'], ax=plt.gca())
plt.title('Sentiment Distribution Across Star Ratings (1-5)')
plt.xlabel('Customer Rating')
plt.ylabel('Number of Reviews')
plt.legend(title='Sentiment')

plt.tight_layout()
plt.show()

# ==========================================
# 5. EXPORT THE RESULTS
# ==========================================
# Save the analyzed data to a new file for your project documentation/presentation
df.to_csv('flipkart_analyzed_sentiment.csv', index=False)
print("\nAnalysis complete! Results saved to 'flipkart_analyzed_sentiment.csv'.")
```
